## Import Pandas for for easy to use dataframes

```python
import pandas as pd
```

## Import Matplotlib to plot the panda-dataframes

```python
import matplotlib.pyplot as plt
```

## Convert .csv file into a usable dataframe

```python
df = pd.read_csv('pctdata.csv')
```

# Slicing dataframes into 3 separate ones
### separating the dataframe since one dataframe has too much data to visualize
### each "id_plane" gets their only dataframe

```python
df0 = df.loc[df['id_plane'] == 0]
df1 = df.loc[df['id_plane'] == 1]
df2 = df.loc[df['id_plane'] == 2]

print("entire df:\nRows: " + str(df.shape[0]) + "\nColumns: " + str(df.shape[1]))
print("\nid_plane 0\nRows: " + str(df0.shape[0]) + "\nColumns: " + str(df0.shape[1]))
print("\nid_plane 1\nRows: " + str(df1.shape[0]) + "\nColumns: " + str(df1.shape[1]))
print("\nid_plane 2\nRows: " + str(df2.shape[0]) + "\nColumns: " + str(df2.shape[1]))
```

# Visualization of chip(id_plane[0-2]) 1 to 3

### Scatter dataframe0

```python
x = df0['id_x']
y = df0['id_y']
fig,ax  = plt.subplots()
ax.scatter(x, y, s=1, color= 'red')
ax.set_title('pct-date of id_plane0', c = '#3282b8', fontsize = 20)
ax.set_xlabel('x', c = '#3282b8', fontsize = 14)
ax.set_ylabel('y', c = '#3282b8', fontsize = 14)
ax.tick_params(colors = '#3282b8')

plt.savefig("MainPlots/pct-date of id_plane0.png", bbox_inches='tight')
plt.show()
plt.close(fig)
```

### Scatter dataframe1

```python
fig,ax  = plt.subplots()
ax.scatter(x, y, s=1, color = 'green')
ax.set_title('pct-date of id_plane1', c = '#3282b8', fontsize = 20)
ax.set_xlabel('x', c = '#3282b8', fontsize = 14)
ax.set_ylabel('y', c = '#3282b8', fontsize = 14)
ax.tick_params(colors = '#3282b8')

plt.savefig("MainPlots/pct-date of id_plane1.png", bbox_inches='tight')
plt.show()
plt.close(fig)
```

## Scatter dataframe2

```python
x = df2['id_x']
y = df2['id_y']
fig,ax  = plt.subplots()
ax.scatter(x, y, s=1, color = 'purple')
ax.set_title('pct-date of id_plane2', c = '#3282b8', fontsize = 20)
ax.set_xlabel('x', c = '#3282b8', fontsize = 14)
ax.set_ylabel('y', c = '#3282b8', fontsize = 14)
ax.tick_params(colors = '#3282b8')

plt.savefig("MainPlots/pct-date of id_plane2.png", bbox_inches='tight')
plt.show()
plt.close(fig)
```

# Beginning thoughts after the first plots
Since the patterns of all the planes are similar, in the way, that the protons hit around the same area on the chips,
leads  me believe, that the chips are placed behind each other and not diagonal or right next to each other.

Also to notice, most of the hits on all three of the chips are close to the center, which adds to the former statement.

## Better visualization by plotting all three chips in a 3D Scatter-Plot
This plot clearly visualizes, that the chips are all placed behind one another.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
fig.set_size_inches(16, 12)

ax.scatter(df0.iloc[::10, 1], 0, df0.iloc[::10, 2], s = 1, color= 'red')
ax.scatter(df1.iloc[::10, 1], 1, df1.iloc[::10, 2], s = 1, color= 'green')
ax.scatter(df2.iloc[::10, 1], 2, df2.iloc[::10, 2], s = 1, color= 'purple')
ax.set_title('pct-data of all chips behind one another', c = '#3282b8', fontsize = 20)
ax.set_xlabel('id_x', fontsize = 14)
ax.set_ylabel('id_plane', fontsize = 14)
ax.set_zlabel('id_y', fontsize = 14)
ax.set_yticks([0, 1, 2])

plt.savefig("MainPlots/pct-data of all chips behind one another.png", bbox_inches='tight')
plt.show()
plt.close(fig)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
fig.set_size_inches(16, 12)

ax.scatter(df0.iloc[::10, 1], 0, df0.iloc[::10, 2], s = 1, color= 'red')
ax.scatter(df1.iloc[::10, 1], 1, df1.iloc[::10, 2], s = 1, color= 'green')
ax.scatter(df2.iloc[::10, 1], 2, df2.iloc[::10, 2], s = 1, color= 'purple')
ax.set_title('pct-data of all chips behind one another', c = '#3282b8', fontsize = 20)
ax.set_xlabel('id_x', fontsize = 14)
ax.set_ylabel('id_plane', fontsize = 14)
ax.set_zlabel('id_y', fontsize = 14)
ax.set_yticks([0, 1, 2])
ax.view_init(20, 60)

plt.savefig("MainPlots/pct-data of all chips behind one another(2).png", bbox_inches='tight')
plt.show()
plt.close(fig)
```

This plot clearly visualizes, that the chips are all placed behind one another

## Show the dense are with the help of a heat map

```python
hm = df[df['id_x'].between(400, 700)]
hm = hm[hm['id_y'].between(200, 400)]

x = hm.iloc[:, 1]
y = hm.iloc[:, 2]

plt.hist2d(x, y, bins = (300,200), cmap = 'plasma')
plt.title('Heatmap of id_x and id_y of the densest area', c = '#3282b8', fontsize = 16)
plt.xlabel('id_x', c = '#3282b8', fontsize = 14)
plt.ylabel('id_y', c = '#3282b8', fontsize = 14)
plt.tick_params(colors = '#3282b8')

plt.savefig("MainPlots/Heatmap of id_x and id_y of the densest area.png", bbox_inches='tight')
plt.colorbar()
plt.show()
```

# Visualize correlation between id_x, id_y and i_event
## Starting of with id_x and all chips

```python
fig, ax  = plt.subplots()
fig.set_size_inches(16, 12)
ax.scatter(df0.iloc[::10, 1], df0.iloc[::10, 3], s=2, color= 'red', label = "Chip0")
ax.scatter(df1.iloc[::10, 1], df1.iloc[::10, 3], s=2, color = 'green', label = "Chip1")
ax.scatter(df2.iloc[::10, 1], df2.iloc[::10, 3], s=2, color = 'purple', label= "Chip2")
ax.set_title('Correlation between id_x and i_event for all chips', c = '#3282b8', fontsize = 30)
ax.legend(loc = "upper right", markerscale = 8, fontsize = 20)
ax.set_xlabel('id_x', c = '#3282b8', fontsize = 28)
ax.set_ylabel('i_event', c = '#3282b8', fontsize = 28)
ax.tick_params(colors = '#3282b8', labelsize = 24)

plt.savefig("MainPlots/Correlation between id_x and i_event for all chips.png", bbox_inches='tight')
plt.show()
plt.close(fig)
```

# Now compare it to id_y and i_event
## All chips

```python
fig, ax  = plt.subplots()
fig.set_size_inches(16, 12)
ax.scatter(df0.iloc[::10, 2], df0.iloc[::10, 3], s=2, color= 'red', label = "Chip0")
ax.scatter(df1.iloc[::10, 2], df1.iloc[::10, 3], s=2, color = 'green', label = "Chip1")
ax.scatter(df2.iloc[::10, 2], df2.iloc[::10, 3], s=2, color = 'purple', label= "Chip2")
ax.set_title('Correlation between id_y and i_event', c = '#3282b8', fontsize = 30)
ax.legend(loc = "upper right", markerscale = 8, fontsize = 20)
ax.set_xlabel('id_y', c = '#3282b8', fontsize = 28)
ax.set_ylabel('i_event', c = '#3282b8', fontsize = 28)
ax.tick_params(colors = '#3282b8', labelsize = 24)

plt.savefig("MainPlots/Correlation between id_y and i_event.png", bbox_inches='tight')
plt.show()
plt.close(fig)
```

You can see larger gaps between some i_events. Those indicate. that the beam going through the chips are not continuous but rather come in bursts.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
fig.set_size_inches(16, 12)

ax.scatter(df0.iloc[:1200, 1], df0.iloc[:1200, 2], df0.iloc[:1200, 3], s = 1, color= 'red', label = 'Chip0')
ax.scatter(df1.iloc[:1200, 1], df1.iloc[:1200, 2], df1.iloc[:1200, 3], s = 1, color= 'green', label = 'Chip1')
ax.scatter(df2.iloc[:1200, 1], df2.iloc[:1200, 2], df2.iloc[:1200, 3], s = 1, color= 'purple', label = 'Chip2')
ax.set_title('Correlation between id_x, id_y and i_event', c = '#3282b8', fontsize = 20)
ax.legend(loc = "upper right", markerscale = 8, fontsize = 20)
ax.set_xlabel('id_x', fontsize = 14)
ax.set_ylabel('id_y', fontsize = 14)
ax.set_zlabel('i_event', fontsize = 14)
ax.set_yticks([0, 1, 2])
ax.view_init(20, 60)

plt.savefig("MainPlots/Correlation between id_x, id_y and i_event.png", bbox_inches='tight')
plt.show()
plt.close(fig)
```

You can see larger gaps between some i_events. Those indicate. that the beam going through the chips are not continuous but rather come in bursts.

## Showing that i_event is a snapshot of a chip, indicating the current hits on the chip and that there can be multiple

```python
fig,ax = plt.subplots()
fig.set_size_inches(14, 8)
ax.hist(df.loc[df['i_event'] == 149, 'id_x'])
ax.set_title('Multiple hits on i_event 149', fontsize = 32, c = '#3282b8')
ax.set_xlabel('id_x', fontsize = 20, c = '#3282b8')
ax.set_ylabel('amount', fontsize = 20, c = '#3282b8')
ax.set_xticks(range(563, 567))
ax.tick_params(colors = '#3282b8', labelsize = 18)

plt.savefig("MainPlots/Multiple hits on i_event 149.png", bbox_inches='tight')
plt.show()
plt.close(fig)

fig,ax = plt.subplots()
fig.set_size_inches(14, 8)
ax.hist(df.loc[df['i_event'] == 149, 'id_y'])
ax.set_title('Multiple hits on i_event 149', fontsize = 32, c = '#3282b8')
ax.set_xlabel('id_x', fontsize = 20, c = '#3282b8')
ax.set_ylabel('amount', fontsize = 20, c = '#3282b8')
ax.set_xticks(range(409, 413))
ax.tick_params(colors = '#3282b8', labelsize = 18)

plt.savefig("MainPlots/Multiple hits on i_event 149(2).png", bbox_inches='tight')

plt.show()
plt.close(fig)
```

Multiple hits during one event indicate multiple proton hits during one given time (i_time_stamp)
