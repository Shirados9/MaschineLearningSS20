```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('pctdata.csv')

df0 = df.loc[df['id_plane'] == 0]
df1 = df.loc[df['id_plane'] == 1]
df2 = df.loc[df['id_plane'] == 2]
```

```python
def showKValue(df, title):
    distortions = []
    K = range(1,7)
    for k in K:
        kmeanModel = KMeans(n_clusters=k)
        kmeanModel.fit(df)
        distortions.append(kmeanModel.inertia_)

    plt.figure(figsize=(16,8))
    plt.plot(K, distortions, '-o')
    plt.title(title, fontsize=24, c = '#3282b8')
    plt.xlabel('k',fontsize=20, c = '#3282b8')
    plt.ylabel('Distortion',fontsize=20, c = '#3282b8')
    plt.xticks(range(1,7), c ='#3282b8', fontsize = 14)
    plt.yticks(c ='#3282b8', fontsize = 14)
    plt.savefig("KMeansPlots/{}.png".format(title), bbox_inches='tight')


    plt.show()
```

```python
showKValue(df, 'K for df')
showKValue(df0, 'K for df0')
showKValue(df1, 'K for df1')
showKValue(df2, 'K for df2')
```

```python
def plotKMeans(df, k, title):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    cluster = kmeans.cluster_centers_
    predict = kmeans.fit_predict(df[['id_x', 'id_y']])
    # df['cluster'] = predict

    color = kmeans.labels_


    plt.scatter(df['id_x'], df['id_y'], s=1, c=color,)
    # plt.scatter(cluster[:,0], cluster[:,1], c='black')
    plt.title(title, fontsize=24, c = '#3282b8')
    plt.xlabel('id_x',fontsize=20, c = '#3282b8')
    plt.ylabel('id_y',fontsize=20, c = '#3282b8')
    plt.tick_params(colors='#3282b8')
    plt.savefig("KMeansPlots/{}.png".format(title), bbox_inches='tight')
    plt.show()
```

```python
plotKMeans(df, 3, 'KMeans clusters for df')
plotKMeans(df0, 3, 'KMeans clusters for df0')
plotKMeans(df1, 3, 'KMeans clusters for df1')
plotKMeans(df2, 3, 'KMeans clusters for df2')

plotKMeans(df, 4, 'KMeans clusters for df')
plotKMeans(df0, 4, 'KMeans clusters for df0')
plotKMeans(df1, 4, 'KMeans clusters for df1')
plotKMeans(df2, 4, 'KMeans clusters for df2')

plotKMeans(df, 2, 'KMeans clusters for df')
plotKMeans(df0, 2, 'KMeans clusters for df0')
plotKMeans(df1, 2, 'KMeans clusters for df1')
plotKMeans(df2, 2, 'KMeans clusters for df2')
```
